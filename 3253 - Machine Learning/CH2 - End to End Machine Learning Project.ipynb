#### Now let's write a function that fetches our data:

The code below creates a directory in our workspace, downloads our .tgz file, and extracts the .csv file from it into the directory

```python
import os
import tarfile
from six.moves import urllib

DL_ROOT = r'https://raw.githubusercontent.com/ageron/handson-ml/master/'
DATA_PATH = os.path.join('datasets','housing')
TGZ_NAME = 'housing.tgz'
DATA_URL = DL_ROOT + 'datasets/housing/' + TGZ_NAME

def fetch_data(data_url, data_path, filename):
    if not os.path.isdir(data_path):
        os.makedirs(data_path)
    tgz_path = os.path.join(data_path, filename)
    urllib.request.urlretrieve(data_url,tgz_path)
    data_tgz = tarfile.open(tgz_path)
    data_tgz.extractall(path=data_path)
    data_tgz.close()

fetch_data(DATA_URL,DATA_PATH,TGZ_NAME)
```

```python
CSV_NAME = 'housing.csv'
def load_data(data_path=DATA_PATH, file_name=CSV_NAME):
    csv_path = os.path.join(data_path,file_name)
    return pd.read_csv(csv_path)
```

### Selecting a performance measure:
Usual suspects are RMSE and MAE:
$$RMSE(X,h) = \sqrt{\frac{1}{m}\sum^{m}_{i=1}(h(x^{(i)})-y^{(i)})^2}$$

$$MAE(X,h) = \frac{1}{m}\sum^{m}_{i=1}|h(x^{(i)})-y^{(i)}|$$

Where:
* $x^{(i)}$ is a vector including the feature values for the $i^{th}$ datapoint
* $y^{(i)}$ is the label value for the $i^{th}$ datapoint

And $X$ is the matrix that includes the transposed feature vectors as such: $$X=\begin{pmatrix}(x^{(1)})^{T} \\ (x^{(2)})^{T} \\ . \\ . \\ . \\ (x^{(m)})^{T}\end{pmatrix}$$

***RMSE*** **is the preferred performance measure**, but may need to use another function such as ***MAE*** if there are many outliers in the data. 
Also note that ***RMSE*** refers to the *Euclidian norm* or the **$\ell_{2}$** norm
And ***MAE*** refers to the *Manhattan norm* or the **$\ell_{1}$** norm (distance between two points via orthogonal lines).

### Notes on how to decide between these two performance measures:
* Do I have outliers in my data? Use **MAE**
* Am I sure they are outliers? Use **MAE**
* Or are they just unexpected values I should still care about? Use **MSE**

## Creating a Test set

To avoid data snooping bias, let's create our test set and put it aside.
This isn't so simple though, a few **considerations**:

Suppose we create our own train test split function:

```python
def split_train_test(data, test_ratio):
    #set seed
    np.random.seed(42)
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

train_set, test_set = split_train_test(housing, 0.2)
print(len(train_set),'train and', len(test_set),'test.')
```

Can we do the same for a train-validate-test split?

```python
def split_train_validate_test(data, validate_ratio, test_ratio):
    np.random.seed(42)
    shuffled_indices = np.random.permutation(len(data))
    test_set_marker = int(len(data) * test_ratio)
    validate_set_marker = int(len(data) * (validate_ratio + test_ratio))
    test_indices = shuffled_indices[:test_set_marker]
    validate_indices = shuffled_indices[test_set_marker:validate_set_marker]
    train_indices = shuffled_indices[validate_set_marker:]
    return data.iloc[train_indices], data.iloc[validate_indices], data.iloc[test_indices]

train_set, validate_set, test_set = split_train_validate_test(housing, 0.2, 0.2)
print(len(train_set),'train,',len(validate_set),'validate and', len(test_set),'test.')
```

Great! Problem is **this will break on an updated dataset**. 
* Perhaps we can use an unique inmutable identifier to decide whether a datapoint goes in the test set. 
    * Could compute a hash of each instance's identifier and use only the last byte of the hash. If this values is lower or equal to 51 (20% of 256) it goes in the test set.
        * Test set will contain 20% of new instances but will not contain any instance that was previously in the training set

```python
import hashlib

def test_set_check(identifier, test_ratio, hash):
    return hash(np.int64(identifier)).digest()[-1]< 256 * test_ratio
```

```python
def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
housing_with_id = housing.reset_index()
```

```python
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'index')

print(len(train_set),'train,', len(test_set),'test.',
      'So',int(len(train_set)/(len(train_set)+len(test_set))*100),'% to train and',
     int(len(test_set)/(len(train_set)+len(test_set))*100),'% to test',
     'out of a total',len(train_set)+len(test_set),'datapoints')
```

```python
def test_set_check(identifier, test_ratio, hash):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio
```

```python
def split_train_validate_test_by_id(data, validate_ratio, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_validate_test_set = ids.apply(lambda id_: test_set_check(id_, (test_ratio + validate_ratio), hash))
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_validate_test_set], data.loc[in_validate_test_set].loc[~in_test_set], data.loc[in_test_set]
```

## Non-random Sampling Methods

Possibility of introducing sampling bias if we use purely random sampling methods if our dataset is not large enough. <br>
Also problematic if original dataset employs *stratified sampling*, population divided into homogeneous subgroups called *strata* and the right number of instancess is sampled from each stratum to ensure the test set is representative of the overall population.

If median income is a very important attribute to predict median house prices, wise to ensure test set is representative of the various categories of incomes in the whole dataset. <br>
**Turn this continuous numerical attribute into categorical**

```python
housing['median_income'].hist(bins=40, figsize=(15,7))
plt.show()
```

Most median income values clustered around $\$$20,000-$\$$50,000, but some go beyond $\$$60,000. Important to ensure sufficient number of instances in dataset for each stratum so that the importance of stratums is not biased

Below, create income category by:
1. Dividing median income by 1.5 (limits number of income categories)
2. Rounding up using ceil (to have discrete categories)
3. And merging all categories above 

```python
housing['income_cat'] = np.ceil(housing['median_income'] / 1.5)
housing['income_cat'].where(housing['income_cat'] < 5, 5.0, inplace=True)
```

### Writing Custom Transformers

To use own tranformer, can create a class and implent *fit(), transform(), fit_transform()*. Can get last one from TransformerMixin as a base class and add BaseEstimator to get *get_params* and *set_params* as well (for automatic hyperparameter tuning).

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

This Transformer class will add the combined attributes we used earlier

```python
rooms_ix, bedroom_ix, population_ix, household_ix = 3, 4, 5, 6
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): #no *args or *kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedroom_ix] / X[: rooms_ix]
            return np.c_[X, rooms_per_household,
                         population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```
