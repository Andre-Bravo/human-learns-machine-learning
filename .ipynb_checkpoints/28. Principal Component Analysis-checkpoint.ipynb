# Principal Component Analysis (PCA)

PCA (also known as a general factor analysis) is a <b>unsupervised</b> statistical technique used to examine the interrelations among a set of variables in order to identify the underlying structure of those variables.

Not used a full machine learning algorithm, but an unsupervised learning algorithm for component reduction

Essentially a transformation of our data, attempting to determine which features explain the most variance in our data

<img src='Python-Data-Science-and-Machine-Learning-Bootcamp/Machine Learning Sections/Principal-Component-Analysis/PCA.png' />

Regression determines a line of best fit to a data set, but factor analysis determines **several** <u>**orthagonal**</u> (at right angles) lines of best fit.
* Which are actually perpendicular to each other in n-dimensional space
    * The n-dimensional space is the variable sample space

Having drawn a line of best fit across data in x and y dimensions, we can add an orthagonal line to better understand the components of the dataset.
* Componets are a linear transformation that chooses a variable system for the data set such that the greatest variance of the data set comes to lie on the first axis
* Likewise, the second greatest variance on the second axis, and so on...
* This process allows us to reduce the number of variables used in an analysis
* We continue this analysis into higher dimensions
* If, we use this technique on a data set with a large number of variables, we can compress the amount of explained variation to just a few components.
* The most challenging part of PCA is interpreting the components.

## Principal Component Analysis with Python

Not used a full machine learning algorithm, but an unsupervised learning algorithm for component reduction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

Principal component analysis: Find what components are the most important ones. That explain the most variance in this data set

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

#### If doing classification on this dataset, we do PCA first to understand what is important in seeing whether a tumor is malignant or bening

Let's use PCA to find the first two principal components, and visualize in this new two-dimensional space

### Scale our data

```python
from sklearn.preprocessing import StandardScaler
```

Scale our data so that each feature has a single unit variance

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

### Perform PCA

```python
from sklearn.decomposition import PCA
```

Let's keep just two principal components and visualize entire dataset using those

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8,6))

plt.scatter(x_pca[:,0],x_pca[:,1])

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
cancer['target_names']
```

```python
np.unique(cancer['target'])
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8,6))

plt.scatter(x_pca[:,0],x_pca[:,1],
            c=cancer['target'],cmap='coolwarm')

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

```python
cancer['target_names']
```

```python
x_pca
```

Components are combinations of underlying data

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```
