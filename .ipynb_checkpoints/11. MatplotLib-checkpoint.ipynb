```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

### Functional method of creating matplotlib plot

```python
plt.plot(x,y,'r-')
plt.xlabel('This is the X')
plt.ylabel('This is the Y')
plt.title('A TITLE')
plt.show()
```

##### Multi-plots on the same canvas

```python
plt.subplot(1,2,1) #Number of rows, number of columns, plot #
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

### Object Oriented Method

##### Create figure objects and call methods of off them

```python
fig = plt.figure() # blank canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #Left, bottom, width, height
                                # percent of canvas we want to occupy
axes.plot(x,y)
axes.set_xlabel('THIS IS X')
axes.set_ylabel('THIS IS Y')
axes.set_title('THIS IS TITLE')
plt.show()
```

### A method with tons of customization

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig1.add_axes([0.15,0.4,0.4,0.4])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
plt.show()
```

### Creating subplots, little less customization

#### You get to set rows and columns

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y, 'r')
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

##### Figure Size and DPI

```python
fig2 = plt.figure(figsize=(8,2)) #size in inches

ax = fig2.add_axes([0,0,1,1])
ax.plot(x,y)

plt.show()
```

```python
fig3,axs = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axs[0].plot(x,y)
axs[0].set_title('Hello')
axes[0].set_xlabel('MY X')
axes[0]

axs[1].plot(y,x)
axs[1].set_title('World')

plt.tight_layout()
plt.show()
```

##### Setting Legends

```python
fig4 = plt.figure(figsize=(8,6))

ax2 = fig4.add_axes([0.1,0.1,0.8,0.8])
ax2.set_title('MY TITLE HERE')

ax2.plot(x,x**2, label='X SQUARED', color='green') #SET LABELS TO AXES FOR LEGEND TO RECOGNIZE
ax2.plot(x,x**3, label='X CUBED', color='blue')

ax2.legend(loc=0) #LOC COMMAND LETS ME CUSTOMIZE LOCATION USING PRESETS, 10 IS CENTRE
                    #CAN ALSO USE TUPLE, TRY loc=(0.1,0.1)

plt.show()
```

### Plot Appearance

##### Line Colour, Line width, Line Style

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#e542f4',linewidth=5) # WILL ACCEPT SIMPLE COLOUR STRINGS (e.g., yellow, green),
                                #OR RGB HEX CODES ('#FF8C00')
axes.plot(x,x**3, linewidth=10, alpha=0.5) #alpha is transparency
axes.plot(x,x**(3.5),linestyle='-.')

plt.show()
```

Markers

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,marker='o')
axes.plot(x,x**3,marker='*',markersize=20,markerfacecolor='white',markeredgewidth=3,markeredgecolor='green')

plt.show()
```

#### Control axis appearance and plot range

```python
fig = plt.figure(figsize=(12,8))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.plot(x,x**3)

axes.set_ylim([0,95])
axes.set_xlim([1,4])
```
