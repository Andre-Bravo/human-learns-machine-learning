#### Hexagonal bin plot

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,figsize = (10,6)) #we increase gridsize as hexagons too small
```
