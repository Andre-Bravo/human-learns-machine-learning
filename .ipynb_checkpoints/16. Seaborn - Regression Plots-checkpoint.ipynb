```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# LM Plot: A simple linear model plot

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

Change formating using matplotlib

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

Affect scatterplot, call argument as in matplotlib <br> NOTE THAT WE USE A DICTIONARY

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':100})
```

#### Use grid instead of separating by plot

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex')
```

#### Adjust size and aspect parameters

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.5, height=8)
```
