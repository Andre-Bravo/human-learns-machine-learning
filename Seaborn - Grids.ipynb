```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

# Pair Grid

Better customization if using pairgrid

```python
sns.PairGrid(iris)
#Note uppercases
```

##### Let's get some scatterplots up. <br>Save the Pair Grid as a variable and map scatter plots unto it

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

# Facet Grid

```python
tips = sns.load_dataset('tips')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')

g.map(sns.distplot, 'total_bill')
```

#### If using the a two variable plot, simply add additional variable

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')

g.map(sns.scatterplot, 'total_bill', 'tip')
```
