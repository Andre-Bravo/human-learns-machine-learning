# K Means Clustering Algorithm

Allows us to cluster unlabelled data in an unsupervised Machine Learning alogrithm
Can be used for:
* Clustering similar Documents
* Clustering Customers based on Features
* Market Segmentation
* Identifying similar physical groups

Divide data into distinct groups such that observations within each gorup are similar
By:
* Choosing a number of Clusters "K"
* Randomly assigning each point to a cluster
* Until clusters stop changing, repeat the following
    * For each cluster, compute the cluster centroid by taking the mean vector of points in the cluster
    * Assign each data point to the cluster for which the centroid is the closest

To choose <b>optimal k</b> value:
* No easy answer
* Can use the elbow method
    * Compute sum of squared error (SSE) for some values of k (for example 2, 4, 6, 8,etc.)
    * The SSE is defined as the sume squared distance between each member of the cluster and its centroid
    
If we plot k against the SSE, note the error *decreases* as k increases. As number of clusters increase, they are smaller, so distortion is also smaller.

Idea of the elbow method is to choose k at the point at which the SSE **decreases abruptly**
* See [here](https://www.datanovia.com/en/lessons/determining-the-optimal-number-of-clusters-3-must-know-methods/) for explanation and other methods

## K Means CLustering with Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

##### Generating artificial data using Scikitlearn

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

Spits out a tuple, first element has our samples and features. Second tuple has out cluster assignments

```python
from sklearn.cluster import KMeans
```

#### K means is unsupervised

```python
kmeans = KMeans(n_clusters=4)
```

## Finidng the optimal k

### 1. Elbow Method

Not that the inertia method is the within cluster sum of squares

```python
Sum_of_squared_distances = []
for k in range(1,20):
    km = KMeans(n_clusters=k)
    km.fit(data[0])
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(range(1,7),Sum_of_squared_distances[:6],'bx-')
```

### 2. Silhouette Method

```python
from sklearn.metrics import silhouette_score
silhoutte_scores = []
for k in range(2,20):
    km = KMeans(n_clusters=k)
    km.fit(data[0])
    centroids = km.cluster_centers_
    score = silhouette_score(data[0],km.labels_,metric='euclidean')
    silhoutte_scores.append(score)
```

```python
plt.xticks(range(2,20,2))

plt.plot(range(2,20),silhoutte_scores,'bx-')
```
