# Categorical plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

Visualization of Groupby action

```python
sns.barplot(x='sex',y='total_bill', data=tips)
```

Shows the result by category of an aggregate function. In this case, we are looking at the mean. Can change estimator

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

Same as barplot, just counts occurrences

## Box (box and whisker) plots and Violin plots

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

Shows quartiles, dots outside are outliers

#### Hue indicator allows us to look at an additional categorical variable such as smoker vs non-smoker

```python
import matplotlib.pyplot as plt
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
#fixing the legend overlap

plt.legend(title='smoker',frameon=False, bbox_to_anchor=(1,1))
#plt.legend(title='smoker',bbox_to_anchor=(1.05,1),loc=2, borderaxespad=0,frameon=False, loc=2)
```

#### Violin plots vs. Box plots

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

Shows kernel density estimation of the underlying distribution. But more difficult to interpret than box

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
#split splits the distributions, lot more info

plt.legend(framealpha=0, loc=2, title='sex', bbox_to_anchor=(1,1))
```

#### Violin gives more info but best for experienced data viz audience, as upposed to management/mainstream audience, use box

# Strip plot

Like a categorical scatterplot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

The jitter variable is set to True by default in newer versions of seaborn. <br>We see the issue if it isn't here, points overlap<br>Try with jitter set to True

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

Again, can add hues

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

##### Note that split has been renamed to dodge in stripplots only it seems

#### Note that harder to interpret vs. box plot and even violin

## Swarm Plot

Similar to strip plot except points adjusted so they do not overlap <br> Better representation of distribution <br> Practically a combination of a strip plot with a violin plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

Drawback: they do not scale well to large numbers, points can get too wide, takes a lot of computation <br> A bit obscure too <br> But could combine with violin plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

##### Messing around with these:

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
sns.swarmplot(x='day',y='total_bill',data=tips, hue='sex',dodge=True, palette='Set2')

plt.legend(title='sex',loc=2,bbox_to_anchor=(1,1))
```

## Factor (Now named Cat) plot

More generalized method for all the graphs we've used this

```python
sns.catplot(x='day', y='total_bill',data=tips, kind='violin')
```
